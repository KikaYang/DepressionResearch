```python
import pandas as pd

e = pd.read_csv("effect_size_by_group.csv")

summary = (e.groupby(["source_dataset", "family_history_flag"])
             .agg(mean_RD=("RD_high_low", "mean"),
                  median_RD=("RD_high_low", "median"),
                  mean_RR=("RR_high_low", "mean"),
                  count=("RD_high_low", "count"))
             .reset_index())

print(summary)
```

In the student dataset, the average risk difference (High–Low financial bucket) is ~0.35, whereas in the professional dataset it is only ~0.04–0.06, suggesting a much stronger association between financial strain and depression among students. Risk ratios fluctuate more in the professional sample due to small baseline (Low) rates, so RD is used as the more stable comparison.
